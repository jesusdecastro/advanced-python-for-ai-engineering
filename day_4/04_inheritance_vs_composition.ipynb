# Día 4: Herencia vs Composición

## Descripción General

En programación orientada a objetos, tenemos dos formas principales de reutilizar código: **herencia** (inheritance) y **composición** (composition). Aunque la herencia es más conocida, la composición suele ser la mejor opción en la mayoría de los casos.

Este notebook explora cuándo usar cada enfoque y por qué el principio "favor composition over inheritance" es tan importante en el diseño de software moderno.

## Objetivos de Aprendizaje

Al finalizar este notebook, serás capaz de:

1. Comprender las diferencias entre herencia y composición
2. Identificar cuándo usar herencia vs composición
3. Aplicar el principio "favor composition over inheritance"
4. Refactorizar jerarquías de herencia a composición
5. Diseñar sistemas flexibles usando composición
