# Día 3: Type Hints Avanzados

## Descripción General

Los type hints básicos nos permiten anotar tipos simples como `int`, `str`, o `list`. Sin embargo, Python ofrece un sistema de tipos mucho más poderoso que incluye tipos genéricos, protocolos, y técnicas de type narrowing. Estos conceptos avanzados nos permiten escribir código más expresivo, reutilizable y seguro.

En este notebook exploraremos cómo usar tipos complejos para modelar estructuras de datos sofisticadas, cómo definir interfaces flexibles con protocolos, y cómo el type checker puede inferir tipos más específicos a través del flujo de control del programa.

## Objetivos de Aprendizaje

Al finalizar este notebook, serás capaz de:

1. Crear clases y funciones genéricas usando `TypeVar` y `Generic` para código reutilizable
2. Definir protocolos para especificar interfaces basadas en estructura (duck typing estático)
3. Usar tipos complejos como `Union`, `Optional`, `Literal`, y `TypedDict` apropiadamente
4. Aplicar técnicas de type narrowing para refinar tipos en tiempo de ejecución
5. Comprender cuándo usar generics vs protocolos para diferentes escenarios de diseño
